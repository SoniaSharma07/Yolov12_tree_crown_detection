```python
from ultralytics import YOLO

model = YOLO('yolov12n.yaml')

# Train the model
results = model.train(
  data='/content/drive/MyDrive/yolov12/data_crown.yaml',
  epochs=5,
  batch=8, # Reduced batch size to mitigate OutOfMemoryError
  imgsz=640,
  scale=0.5,  # S:0.9; M:0.9; L:0.9; X:0.9
  mosaic=1.0,
  mixup=0.0,  # S:0.05; M:0.15; L:0.15; X:0.2
  copy_paste=0.1,  # S:0.15; M:0.4; L:0.5; X:0.6
  device="0",
)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Load model
model = YOLO("/content/drive/MyDrive/yolov12/yolov12/runs/detect/train/weights/best.pt")

# Run inference with low confidence threshold
results = model("/content/drive/MyDrive/yolov12/test_2.jpg", conf=0.5)

# Plot with customization
img_with_boxes = results[0].plot(
    line_width=1,
    font_size=0.5,
    labels=False
)

# Convert BGR to RGB for matplotlib
img_rgb = img_with_boxes[..., ::-1]

# Display using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("YOLOv12 Detection")
plt.show()
```
